# Controlling Figure Aesthetics in Seaborn

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure()
x1 = [10,20,5,40,0]
x2 = [30,43,9,7,20]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set()
plt.figure()
x1 = [10,20,5,40,0]
x2 = [30,43,9,7,20]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure()
x1 = [10,20,5,40,0]
x2 = [30,43,9,7,20]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
sns.set()
plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]
with sns.axes_style('dark'):
    plt.plot(x1, label = "Group A")
    plt.plot(x2, label = "Group B")
plt.legend()
plt.show()
```

```python
sns.set_context("poster")
plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]
plt.plot(x1, label = "Group A")
plt.plot(x2, label = "Group B")
plt.legend()
plt.show()
```

# Example in Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Mydata = pd.read_csv("scores.csv")
Mydata.head()
```

```python
Group_A = Mydata[Mydata.columns[0]].tolist()
Group_B = Mydata[Mydata.columns[1]].tolist()
Group_C = Mydata[Mydata.columns[2]].tolist()
Group_D = Mydata[Mydata.columns[3]].tolist()
```

```python
print(Group_A)
```

```python
print(Group_B)
```

```python
print(Group_C)
```

```python
print(Group_D)
```

```python
data = pd.DataFrame({'Groups':['Group A']*len(Group_A)+['Group B']*len(Group_B)
                     +['Group C']*len(Group_C)+['Group D']*len(Group_D),
                    'IQ Score': Group_A+Group_B+Group_C+Group_D})
```

```python
sns.set_context("paper")
plt.figure(dpi = 100)
sns.set_style('whitegrid')
sns.boxplot(x='Groups', y='IQ Score', data=data)
sns.despine(left=True, right=True, top=True)
plt.title("IQ Scores for Different Test Groups")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
example_data = pd.read_csv("flight_details.csv")
```

```python
Data_Frame = example_data.pivot(index="Months", columns="Years", values="Passengers")
```

```python
sns.set()
plt.figure(dpi=150)
sns.heatmap(Data_Frame, cmap=sns.light_palette("orange", as_cmap=True, reverse=True))
plt.title("Flight Passengers from 2001 to 2012")
plt.show()
```

# Basic Plots in Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
example_data = pd.read_csv("salary.csv")
```

```python
sns.set(style="whitegrid")
sns.barplot(x="Education", y="Salary", hue="District",data=example_data)
```

```python
sns.barplot(x="Education", y="Salary", hue="Education", data=example_data, palette="pastel")
```

```python
x = np.random.normal(size=50)
sns.distplot(x)
```

```python
sns.kdeplot(x, fill=True)
```

```python
example_data = pd.read_csv("salary.csv")
sns.set(style="white")
sns.jointplot(x="Salary", y="Age", data=example_data)
```

```python
sns.jointplot(x="Salary", y="Age", data=example_data, kind='kde', xlim=(0, 500000), ylim=(0,100))
```

```python
mydata = pd.read_csv("basic_details.csv")
sns.set(style="ticks", color_codes=True)
g=sns.pairplot(mydata, hue="Groups")
```

```python
sns.set(style="whitegrid")
sns.violinplot(x="Education", y="Salary", hue="Gender", data=example_data, split=True, cut=0)
```

```python
example_1_data = pd.read_csv("movie_scores.csv")
movie_score = pd.DataFrame({"Movie Title": list(example_1_data["MovieTitle"]) * 2,
                           "Score": list(example_1_data["AudienceScore"]) + 
                            list(example_1_data["Tomatometer"]),
                           "Type": ["AudienceScore"] * len(example_1_data["AudienceScore"])
                           + ["Tomatometer"] * len(example_1_data["Tomatometer"])})
sns.set()
plt.figure(figsize=(10, 5), dpi=300)
ax = sns.barplot(x="Movie Title", y="Score", hue="Type", data = movie_score)
plt.xticks(rotation=10)
plt.title("Movies Scores Comparison")
plt.xlabel("Movies")
plt.ylabel("Scores")
plt.show()
```
