# Controlling Figure Aesthetics in Seaborn

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure()
x1 = [10,20,5,40,0]
x2 = [30,43,9,7,20]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set()
plt.figure()
x1 = [10,20,5,40,0]
x2 = [30,43,9,7,20]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure()
x1 = [10,20,5,40,0]
x2 = [30,43,9,7,20]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
sns.set()
plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]
with sns.axes_style('dark'):
    plt.plot(x1, label = "Group A")
    plt.plot(x2, label = "Group B")
plt.legend()
plt.show()
```

```python
sns.set_context("poster")
plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]
plt.plot(x1, label = "Group A")
plt.plot(x2, label = "Group B")
plt.legend()
plt.show()
```

# Example in Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Mydata = pd.read_csv("scores.csv")
Mydata.head()
```

```python
Group_A = Mydata[Mydata.columns[0]].tolist()
Group_B = Mydata[Mydata.columns[1]].tolist()
Group_C = Mydata[Mydata.columns[2]].tolist()
Group_D = Mydata[Mydata.columns[3]].tolist()
```

```python
print(Group_A)
```

```python
print(Group_B)
```

```python
print(Group_C)
```

```python
print(Group_D)
```

```python
data = pd.DataFrame({'Groups':['Group A']*len(Group_A)+['Group B']*len(Group_B)
                     +['Group C']*len(Group_C)+['Group D']*len(Group_D),
                    'IQ Score': Group_A+Group_B+Group_C+Group_D})
```

```python
sns.set_context("paper")
plt.figure(dpi = 100)
sns.set_style('whitegrid')
sns.boxplot(x='Groups', y='IQ Score', data=data)
sns.despine(left=True, right=True, top=True)
plt.title("IQ Scores for Different Test Groups")
plt.show()
```
